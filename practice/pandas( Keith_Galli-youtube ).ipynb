## Filtering Data

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')] 
# All that records that have Type 1 = Grass and Type 2 = Poison
```

```python
df.loc[df['Name'].str.contains('Mega')] # Records that have 'Mega' in their name
```

```python
df.loc[~df['Name'].str.contains('Mega')] # Records that doesn't have 'Mega' in their name
```

### Using Regular Expression

```python
import re
df.loc[df['Type 1'].str.contains('Grass|Fire' , regex=True)]
```

### Using Regular Expression but the characters doesn't matter here

```python
import re
df.loc[df['Type 1'].str.contains('grass|fire' ,flags=re.I, regex=True)]
```
